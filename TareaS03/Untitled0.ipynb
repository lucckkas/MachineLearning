```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = np.linspace(0, 40, 1000) # start, finish, n points
```

Problema:

Dado que la compra de 2 boletos de adulto y 1 boleto de niño tiene un costo de  \\$32, y la compra de 1 boleto de adulto y 3 boletos de niño tiene un costo de \\$36, se requiere encontrar el precio individual de cada tipo de boleto.

Las variables del problema son el precio de cada tipo, llamaremos A al precio de un boleto de adulto y N a un boleto de niño. Con esto armamos las ecuaciones que nos da el enunciado:

&nbsp;&nbsp;&nbsp;&nbsp; 1) $2A + 1N = 32$

&nbsp;&nbsp;&nbsp;&nbsp; 2) $1A + 3N = 36$

Las reordenamos y queda:


```python
d_r = (32-N)/2  # nombre sacado del ejemplo, no significa nada
d_s = 36-3*N    # nombre sacado del ejemplo, no significa nada
```

Ahora graficamos las funciones:

```python
fig, ax = plt.subplots()
plt.title('precio')
plt.xlabel('niño')
plt.ylabel('adulto')
ax.set_xlim([0, 35])
ax.set_ylim([0, 40])
ax.plot(N, d_r, c='green')  # 2 adultos 1 niño
ax.plot(N, d_s, c='brown')  # 1 adulto  3 niños
plt.axvline(x=8, color='purple', linestyle='--')
_ = plt.axhline(y=12, color='purple', linestyle='--')
```

Podemos ver como se intersectan las rectas en el punto (8,12) por lo que el valor de los boletos de adulto es de \$12 y el de los niños es de \$8
